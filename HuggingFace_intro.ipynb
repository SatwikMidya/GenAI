```python
from transformers import pipeline
```

```python
'''
Text Classification - >
Sentiment Analysis
Topic Classification
Spam Detection
Language Identification
Assigning a category to a piece of text
'''

classifier = pipeline('text-classification')

'''
Token Classification ->
Named entity recognition
part of speech tagging
Assigning labels to individual tokens based on their context
'''

token_classifier = pipeline('token-classification')

'''
Question answering ->
'''

question_answerer = pipeline('question-answering')

'''
Summarization -> Condensing long documents into shorter summaries
'''

summarizer = pipeline('summarization')

'''
Translation -> Translating text from one language to another
'''

translator = pipeline('translation', model="Helsinki-NLP/opus-mt-en-es")

'''
Feature extraction -> Extracting numerical representations of text
'''

feature_extractor = pipeline('feature-extraction')

'''
Text2Text Generation -> General-purpose text transformation (includes summarization and translation)
'''

text2text_generator = pipeline('text2text-generation')

'''
Fill-mask -> Filling in the blanks in a given text
'''

fill_mask = pipeline('fill-mask')



'''
Image Classification -> Classifying images into different categories
'''

image_classifier = pipeline('image-classification')

'''
Image Segmentation -> Segmenting different parts of an image into classes
'''

image_segmenter = pipeline('image-segmentation')

'''
Object Detection -> Identifying and localizing objects in an image
'''

object_detector = pipeline('object-detection')


# Speech processing task

'''
Converting Spoken Language into text.
'''

speech_recognizer = pipeline('automatic-speech-recognition')

```

## **Tokenization**

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, DistilBertTokenizer, DistilBertForSequenceClassification

model_name1 = "distilbert-base-uncased-finetuned-sst-2-english"
model_name2 = "nlptown/bert-base-multilingual-uncased-sentiment"
mytokenizer1 = DistilBertTokenizer.from_pretrained(model_name1)
mytokenizer2 = AutoTokenizer.from_pretrained(model_name2)
mymodel1 = DistilBertForSequenceClassification.from_pretrained(model_name1)
mymodel2 = AutoModelForSequenceClassification.from_pretrained(model_name2)

classifier = pipeline("sentiment-analysis", model = mymodel1, tokenizer = mytokenizer1)
res = classifier("I am happy to learn transformers")
print(res)

classifier2 = pipeline("sentiment-analysis", model = mymodel2, tokenizer = mytokenizer2)
res2 = classifier("I am happy to learn transformers")
print(res2)
```
